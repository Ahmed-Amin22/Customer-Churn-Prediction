```python
import pandas as pd
from tensorflow import keras
from tensorflow.keras.models import Sequential , Model
from tensorflow.keras.layers import Dense , Input
from tensorflow.keras.optimizers import Adam , SGD
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('/content/Telco_customer_churn.xlsx')
data.head()
```

```python
def data_info(data):
  cols=[]
  dtypes=[]
  n_unique=[]
  unique_values=[]
  nulls=[]

  for col in data.columns:
    cols.append(col)
    dtypes.append(data[col].dtype)
    n_unique.append(data[col].nunique())
    unique_values.append(data[col].unique())
    nulls.append(data[col].isna().sum())

  return pd.DataFrame({
      'Column':cols,
      'Dtype':dtypes,
      'N Unique':n_unique,
      'Unique Values':unique_values,
      'Nulls':nulls
    })
```

```python
data_info(data)
```

```python
data.drop(['CustomerID',	'Count',	'Country',	'State',	'City',	'Zip Code',	'Lat Long','Churn Label'], axis=1 , inplace=True)
```

```python
data_info(data)
```

```python
le =LabelEncoder()
data['Contract'] = le.fit_transform(data['Contract'])
```

```python
data['Churn Value'].value_counts()
```

```python
data['Churn Reason'].value_counts()
```

```python
data['Churn Reason'].fillna('not leave', inplace=True)
```

```python
data_info(data)
```

```python
data['Total Charges'] = pd.to_numeric(data['Total Charges'] , errors='coerce')
```

```python
data['Total Charges'].isna().sum()
```

```python
data['Total Charges'].fillna(data['Total Charges'].mean() , inplace=True)
```

```python
data['Total Charges'].isna().sum()
```

```python
data = pd.get_dummies(data , drop_first=True)
```

```python
data.head()
```

```python
x=data.drop('Churn Value' , axis=1)
y=data['Churn Value']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3 , random_state=42)
```

```python
sd = StandardScaler()
x_train=sd.fit_transform(x_train)
x_test=sd.transform(x_test)
```

```python
model = Sequential()
model.add(Dense(50 , activation= 'relu' , input_shape=(x_train.shape[1],)))
model.add(Dense(50 , activation= 'relu' ))
model.add(Dense(1 , activation= 'sigmoid' ))
model.summary()
```

```python
opt=Adam(learning_rate=0.001 , beta_1=0.9 , beta_2 = 0.99)
```

```python
model.compile(loss = 'binary_crossentropy' , optimizer= opt , metrics = ['accuracy'])
```

```python
history = model.fit(x_train , y_train ,validation_split=0.1 , epochs=100 , batch_size= 8)
```

```python
plt.plot(history.history['accuracy'] , label = 'train')
plt.plot(history.history['val_accuracy'] , label = 'validation')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'] , label = 'train')
plt.plot(history.history['val_loss'] , label = 'validation')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred_labels = (y_pred > 0.5).astype(int)
```

```python
y_pred_labels
```

```python
cm = confusion_matrix(y_test , y_pred_labels)
print(cm)
```

```python
sns.heatmap(cm , annot=True , fmt='d' , cmap='Blues', xticklabels=['class 0' , 'class 1'] , yticklabels=['class 0' , 'class 1'])
plt.xlabel('Predict label')
plt.ylabel('true label')
plt.title('confusion matrix')
plt.show()
```

### Early Stopping

```python
model = Sequential()
model.add(Dense(50 , activation= 'relu' , input_shape=(x_train.shape[1],)))
model.add(Dense(50 , activation= 'relu' ))
model.add(Dense(1 , activation= 'sigmoid' ))
model.summary()
```

```python
opt=Adam(learning_rate=0.001 , beta_1=0.9 , beta_2 = 0.99)
```

```python
model.compile(loss = 'binary_crossentropy' , optimizer= opt , metrics = ['accuracy'])
```

```python
early_stopping= EarlyStopping(monitor='accuracy' , patience=5)
```

```python
history= model.fit(x_train , y_train ,validation_split=0.1 , epochs=100 , batch_size= 8 , callbacks=[early_stopping])
```
